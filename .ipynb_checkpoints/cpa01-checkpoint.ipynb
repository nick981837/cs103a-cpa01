```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('fossil.csv')
df
```

```python
df.columns
```

```python
df.index
```

```python
df2 = df[df['Country']=='TAIWAN']
df2['Total'].plot()
```

```python
column = 'Total'

df3 = pd.pivot_table(df, values=column, index=["Year"], columns=["Country"])
df3
```

```python
df3[['UNITED KINGDOM','TAIWAN', 'CANADA']].plot(figsize=(15,10))
plt.grid()
plt.title("CO2 EMISSIONS FROM FOSSIL FUELS",fontsize=32)
```

```python
df4 = df.groupby(['Year']).Total.agg(sum)
df4.plot(figsize=(20,10))
plt.grid()
```

##Discussion
### 1.Why the CO2 Emissions from Fossil Fuels has beend sharply decreasing between 1900s to 1950s in United Kindom?
#### The most significant factors include a cleaner electricity mix based on gas and renewables instead of coal, as well as falling demand for energy across homes, businesses and industry.


### 2.What caused the CO2 emissions from Fossol Fuels has been sharply increasing between 1950s to 2000s in Taiwan?
#### High-speed growth accompanied by quick industrialization began in the late-1950s. Taiwan became known for its cheap manufactured exports produced by small enterprises bound together by flexible sub-contracting networks.
