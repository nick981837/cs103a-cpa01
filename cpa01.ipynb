```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('fossil.csv')
df
```

```python
df.columns
```

```python
df.index.tolist()
```

```python
df2 = df[df['Country']=='TAIWAN']  # select out just those rows with iso_code = USA
df2['Total'].plot()
```

```python
column = 'Total'

df3 = pd.pivot_table(df, values=column, index=["Year"], columns=["Country"])
df3
```

```python
df3[['UNITED KINGDOM','TAIWAN', 'CANADA']].plot(figsize=(15,10))
plt.grid()
plt.title(column,fontsize=32)
```

```python
df4 = df.groupby(['Year']).Total.agg(sum)
df4.plot(figsize=(20,10))
plt.grid()
```
